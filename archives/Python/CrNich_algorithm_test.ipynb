```python
from numpy import *
from matplotlib.pyplot import *
```

```python
def delta(i,j):
    if i==j:
        return 1
    return 0

def H(ax,ay,axy,x,y):
    mat = zeros( (x.size, y.size) )
    for i in range(x.size):
        for j in range(y.size):
            value = (
                axy * delta(i-1,j-1) -
                (ay + 2*axy) * delta(i,j-1)   +
                axy * delta(i+1,j-1) -
                
                (ax + 2*axy) * delta(i-1,j) +
                (4*axy + 2*ax + 2*ay) * delta(i,j)   -
                (ax + 2*axy) * delta(i+1,j) +
                
                axy * delta(i-1,j+1) -
                (ay + 2*axy) * delta(i,j+1)   +
                axy * delta(i+1,j+1)
            )
            mat[i,j] = value
    return mat

def Tarray(T0,Hmat,tf=10,dt=0.01,c=1):
    t = arange(0,tf,dt)
    T = zeros( (t.size, Hmat.size), dtype=complex )
    T[0] = T0
    
    for i in range(1,t.size):
        T[i] = matmul(Hmat,T[i-1]) + c*dt
        
    return t,T
```

```python
dt = 0.01
dx = 0.01
dy = 0.01
a = 0.6

ax = a*dt/dx**2
ay = a*dt/dy**2
axy = (a*dt)**2/(dx**2 * dy**2)

x = arange(-0.1,0.1,dx)
y = arange(-0.1,0.1,dx)

Hmat = H(ax,ay,axy,x,y)
```
